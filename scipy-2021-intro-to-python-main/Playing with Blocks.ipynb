```python
def invert_color(rgb_triple):
    red = 255 - rgb_triple[0]
    green = 255 - rgb_triple[1]
    blue = 255 - rgb_triple[2]
    return(red, green, blue)
```

```python
invert_color(colors.Blue)
```

```python
invert_color(grid[10,12].rgb)
```

```python
def invert_grid_color(input_grid):
    new_grid = BlockGrid(input_grid.height, input_grid.width)
    
    for row in range(input_grid.height):
        for column in range(input_grid.width):
            new_grid[row, column].red = 255 - input_grid[row,column].red
            new_grid[row, column].green = 255 - input_grid[row,column].green
            new_grid[row, column].blue = 255 - input_grid[row,column].blue
            
    return new_grid
```

```python
grid.show()
invert_grid_color(grid).show()
```

```python
def invert_grid_colors2(input_grid):
    new_grid = input_grid.copy()
    for new_block, input_block in zip(new_grid, input_grid):
        new_block.rgb = invert_color(input_block.rgb)
    return new_grid
```

```python
grid.show()
invert_grid_colors2(grid).show()
```
