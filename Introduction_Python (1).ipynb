```python
# here I'm making two 3x3 arrays, one with even numbers and the other with odd
ta3e = np.array(([0,2,4],[6,8,10],[12,14,16]))
ta3o = np.array(([1,3,5],[7,9,11],[13,15,17]))
# each array exists independently of anything else, but I can also create an array 
# from their combination
```

```python
taeo = np.array((ta3e,ta3o))
# now all of the data that was stored in 2 different arrays, 
# can be stored and accessed from one object
taeo  # typing the name of an array will display its components
```

```python
# now, the first index (0 or 1) will determine whether I'm accessing the even list or the odd list respectively
# the second index (0,1,2) will determine which row I'm selecting in the even/odd lists
# the third index (0,1,2) will determine which column in the row

# so, if I want to access the even list, third row, second number, 
# my indices should be 0,2,1 like so: 
taeo[0][2][1]
```

```python
# or the odd list, first row, third number, my indices should be 1,0,2
taeo[1][0][2]
```

```python
# I can always view the shape of a multidimensional numpy array, which will give me its dimensions
taeo.shape  # I have constructed this as two 3x3 arrays
```

```python
# in order to explore another way of sifting through our data, I am going to 
# recast our multidimensional array into a 1D list
# the list has 18 elements (2x3x3)
size = 1 # this is instantiting the variable that's going to hold the size of the 1D list
# the following is a "for" loop; the structure of a for loop is 
# for x in y: do this
# you supply the x, which is the counting variable you are iterating over (in this case the dimensions
# of each axes in the array, so dim takes on the values 2 for the first time through the loop, 
# 3 for the second time in the loop, and 3 again for the third time through the loop); 
# since my number of dimensions is 3, this loop will execute 3 times
# each time through the loop, I take the variable size multiply it by each of the array's dimensions
# so the first time through the loop, size = 2; the second time through the loop, size = 2*3, and the third time 
# through the loop size = 2*3*3
# writing size *= dim is a shorthand for the equation size = dim*size

for dim in np.shape(taeo): size *= dim  # the size is equal to the multiple of each
                                        # dimension
size  # finally the output of size is just the total number of elements in our 2x3x3 array
```

```python
# now I'm defining a 1D list with the same number of elements as our 3D array
# so that I can store all of the data into this 1D list
list_test = np.reshape(taeo,size) # the reshape function takes a multidimensional array
                                    # and recasts it into a 1D list of length size
```

```python
list_test # here is what my 1D list looks like; all the same information, but now the container 
            # is a 1D list instead of a 3D array
```

```python
# now I am going to show you a method for sifting through numpy arrays which only requires
# iterating through the data once
# we can do a search through the data looking for, say, all the even numbers.
# then instead of storing those numbers themselves, we instead use a mask that 
# stores the location of the numbers 
list_index=np.zeros(len(list_test)) # this is instantiating the array that will store the a mask that contains
                                    # the locations of the data we want to keep from our data array
list_indexe=np.zeros(len(list_test)) # here will store our even number mask
list_indexo=np.zeros(len(list_test)) # here will store our odd number mask
even = np.zeros(len(list_test)) # here will store our even numbers
odd = np.zeros(len(list_test)) # here will store our odd numbers
```

```python
for i in range(len(list_test)): # another for loop, this time iterating over the length of our 1D array
                                # i is just a counter that starts at 0 and goes to 17
    if (list_test[i]%2==0): # an "if" statement, which is a conditional that only executes when the condition is true
                            # using % is a modulus operator, which computes the remainder
                            # of a division by 2; if the remainder is equal to zero, then the number 
                            # is even
        list_index[i] = 1   # so in that case, I make the list index 1, otherwise it's zero
```

```python
list_indexe = list_index # now list_index contains a mask of sorts; it's a 1 everywhere in our data list
            # where there was an even number, and its a zero for odd
list_indexe
    
```

```python
# here I'm creating an inverted mask for the odd numbers
for i in range (len(list_indexe)): 
    c = list_indexe[i]   # always create a dummy variable for temporarily storing array data
                        # this will prevent you from modifying the original array
    if c == 0: list_indexo[i]=1  # take all of the zero elements in list_indexe and replace them with 1
    if c == 1: list_indexo[i]=0  # take all of the one elements in the list_indexe and replace them with 0
#list_indexo = list_indexe
list_indexo
```

```python
# now, instead having to loop over the entire array every time I want to do something
# with only the even numbers, I can multiply the original array by my mask, and the only 
# elements remaining will be those I selected on (the even ones in this case)
# this can be useful if I ever need to iterate through my data looking for specific particles
#(like following them over time) rather than sifting through all of the data at every timestep
#(which can be hundreds of thousands of points)
#list_indexe[1]*list_test[1]
even=list_indexe*list_test
# and the odd is just the opposite
odd=list_indexo*list_test
```

```python
# the purpose of the above exercise was to show you how we create places to store data, 
# how we reference, move, and manipulate that data
# now we will create and visualize a simulated 2dimensional data set
```

```python
res = 5000  # here I am defining the number of points in my simulation
test_gauss1 = np.zeros(res)  # here I am instantiating the 1D arrays (a list) that will 
test_gauss2 = np.zeros(res)  # hold 2 sets of my simulation data

# this is now using a gaussian normal distribution in order to draw random numbers; 
# so the numbers aren't truly random
# random.normal is the function from the random library that will provide our simulated 
# data
# the great thing about simulated data, is that if the distributions in your data are
# the same, or even similar, you can always create simulated data on which to test
# out your hypotheses every time 
# to execute this cell's commands, you will recreate a new data set, which 
# has the same parameters as the previous one, but is taken from a new random sample

for i in range(0,res): # this is a for loop that repeats all of the commands inside for 
                        # res-0 times.  i is the counter, and increments by 1 each time 
                        # through the loop; "range" is a function that outputs the interval 
                        # between 0 and res; so 5000 in this case; we're iterating through
                        # this loop 5000 times
                        # i begins with i=0, so the index [i] starts off with the beginning 
                        # of our list
# normal is a function within random that will take a random draw from a 
# gaussian distribution of points centered at zero with different variances
# both gaussians are centered on zero (loc=0)    
# the first gaussian is broad (scale=1.5)
    test_gauss1[i] = np.random.normal(loc=0,scale=1.5)
# the second gaussian is narrow (scale 0.5)
    test_gauss2[i] = np.random.normal(loc=0,scale=0.5)
```

```python
# now for plotting
# the data in test_gauss1 and test_gauss2 are just values centered around zero 
# with a given variance, so one way to present the data for each is to bin it  
# for a histogram type plot
plot.figure()   # this is a function to instatiate your figure object which will contain the plot

H1 = plot.hist(test_gauss1)  # plot.hist is for plotting histogram data
H2 = plot.hist(test_gauss2)  # again, I set the plot equal to dummy variables H1 and H2 so 
# that if I ever have to work with that data, I do not change the original data set while I'm working
# by default, these plots will be in different colors
```

```python
# we can also take this data and use it to plot a 2D histogram plot
# where our guassian points represent the x and y coordinates on the plot
# the color reflects the height of the histogram at the xy position
plot.figure(figsize=(6,2)) # this is to ensure that the dimensions of the plot
                            # reflect the actual spatial dimensions
H2d = plot.hist2d(test_gauss1,test_gauss2)
```

```python
# alternatively, we can produce a number line plot, 
# where the 1D point density should resemble a gaussian
# now that we have 2d data (gaussian in x and y), we can also do a scatter plot
plot.xlim(-5,5); plot.ylim(-5,5)  # these set the xy-axes limits
S2d = plot.scatter(test_gauss1,test_gauss2, s=1,c="red") # s=1 refers to the point size

# the resulting distribution of points has a gaussian shape in both x and y, 
# where the y-coordinates are the narrower distribution
```

```python
# now interchanging my distributions and plotting both in different colors
test_gauss3 = np.zeros(res) # creating empty lists filled with zeros
test_gauss4 = np.zeros(res)


for i in range(0,res):
    test_gauss4[i] = np.random.normal(loc=0,scale=1.5)
    test_gauss3[i] = np.random.normal(loc=0,scale=0.5)
    
```

```python
plot.xlim(-5,5); plot.ylim(-5,5) # here I am manually setting my xy-axes limits
S2d = plot.scatter(test_gauss1,test_gauss2, s=1,c="red") 
S2db = plot.scatter(test_gauss3,test_gauss4, s=1, c="blue")

# this visualization shows two distributions of simulated data we've created
# the distributions are identically programmed, and so should look more or less the same
# except that one is rotated 90 degrees from the other
# notice that, despite our axes being programmed identically, the plot below may 
# not look square
```
